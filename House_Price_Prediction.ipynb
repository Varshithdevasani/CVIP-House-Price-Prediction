```python
df['POSTED_BY'].fillna(df['POSTED_BY'].mode()[0],inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
from collections import Counter as count
le=LabelEncoder()
```

```python
print('before label encoding',count(df['POSTED_BY']))
df['POSTED_BY']=le.fit_transform(df['POSTED_BY'])
print('after replace',count(df['POSTED_BY']))
```

```python
print('before label encoding',count(df[ 'BHK_OR_RK']))
df[ 'BHK_OR_RK']=le.fit_transform(df[ 'BHK_OR_RK'])
print('after replace',count(df[ 'BHK_OR_RK']))
```

```python
x=df[['POSTED_BY', 'UNDER_CONSTRUCTION', 'RERA', 'BHK_NO.', 'BHK_OR_RK',
       'SQUARE_FT', 'READY_TO_MOVE', 'RESALE', 'LONGITUDE',
       'LATITUDE']]
```

```python
y=df['PRICE']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
ln=LinearRegression()
```

```python
ln.fit(x_train,y_train)
```

```python
train_score=round(ln.score(x_train,y_train)*100,2)
print('linear regression train score:',train_score)
```

```python
coeff_data=pd.DataFrame(ln.coef_,x.columns,columns=['Coefficient'])
```

```python
coeff_data
```

```python
y_pred_test = ln.predict(x_test)
y_pred_train = ln.predict(x_train)
```

```python
plt.scatter(y_test,y_pred_test)
```

```python
sns.histplot((y_test-y_pred_test),bins=20);
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
mse = mean_squared_error(y_test,y_pred_test)
print("Mean Squared Error for test:", mse)
```

```python
mse_t= mean_squared_error(y_train,y_pred_train)
print("Mean Squared Error for train:", mse_t)
```

```python
mae = mean_absolute_error(y_test,y_pred_test)
print("Mean Absolute Error:", mae)
```

```python
r2 = r2_score(y_test,y_pred_test)
print("R-squared Score:", r2)
```

```python
np.sqrt(mean_squared_error(y_test,y_pred_test))
```
